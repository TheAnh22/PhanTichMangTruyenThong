```python
post_df_full1=post_df_full

post_df_full1['reactions'] = post_df_full1['reactions'].apply(lambda x : dict(eval(x)) )
post_df_full_reactions = post_df_full1['reactions'].apply(pd.Series )
# Merge cột vào dataframe
post_df_full_with_reactions = pd.concat([post_df_full1, post_df_full_reactions], axis=1).drop('reactions', axis=1)
# Sau khi thực hiện các bước ban đầu, như đã mô tả trong mã code ban đầu, thêm dòng sau:
post_df_full_with_reactions.fillna(0, inplace=True)  # Thay thế NaN bằng 0
columns_to_convert_to_int = ['thích', 'yêu thích', 'haha', 'wow', 'buồn', 'phẫn nộ', 'thương thương', 'shares', 'comments', 'reaction_count']

# Chuyển đổi các cột sang kiểu int và loại bỏ số thập phân
post_df_full_with_reactions[columns_to_convert_to_int] = post_df_full_with_reactions[columns_to_convert_to_int].astype(int)


post_df_full_with_reactions[['post_text','thích','yêu thích','haha','wow','buồn','phẫn nộ','thương thương',
                                 'shares','comments','reaction_count']]
```

```python
post_df_full_with_reactions=post_df_full_with_reactions.sort_values(by='reaction_count', ascending=False) # Sap xep reac giam dan`

post_df_full_with_reactions[['post_text','thích','yêu thích','haha','wow','buồn','phẫn nộ','thương thương',
                                 'shares','comments','reaction_count']]
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(post_df_full_with_reactions['time'], post_df_full_with_reactions['thích'], label = "Like", marker="o")
ax.plot(post_df_full_with_reactions['time'], post_df_full_with_reactions['yêu thích'], label = "Love", marker="o")
ax.plot(post_df_full_with_reactions['time'], post_df_full_with_reactions['haha'], label = "Ha Ha", marker="o")
ax.plot(post_df_full_with_reactions['time'], post_df_full_with_reactions['wow'], label = "Wow", marker="o")
ax.plot(post_df_full_with_reactions['time'], post_df_full_with_reactions['buồn'], label = "Sad", marker="o")
ax.plot(post_df_full_with_reactions['time'], post_df_full_with_reactions['phẫn nộ'], label = "Angry", marker="o")
ax.plot(post_df_full_with_reactions['time'], post_df_full_with_reactions['thương thương'], label = "Care", marker="o")
ax.plot(post_df_full_with_reactions['time'], post_df_full_with_reactions['shares'], label = "Shares", marker="s")
ax.plot(post_df_full_with_reactions['time'], post_df_full_with_reactions['comments'], label = "Comments", marker="*")
ax.plot(post_df_full_with_reactions['time'], post_df_full_with_reactions['comments'], label = "reaction_count", marker="*")

plt.legend()
from matplotlib.dates import DateFormatter
ax.xaxis.set_major_formatter(DateFormatter('%d-%m-%y %H:%M'))
```

```python
#Xuat ra tu duoc dung nhieu nhat
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

# Đọc tệp văn bản và chuyển đổi nội dung thành một chuỗi
with open('/content/drive/MyDrive/MXH/res.txt', 'r') as file:
    text = file.read()

# Loại bỏ các ký tự không cần thiết (ví dụ: dấu câu)
text = re.sub(r'[^\w\s]', '', text)

# Chia văn bản thành các từ riêng lẻ
words = text.split()

# Đếm tần suất xuất hiện của các từ
word_counts = Counter(words)

# Lấy 20 từ được sử dụng nhiều nhất
top_50_words = word_counts.most_common(50)

# Tạo danh sách các từ và tần suất
word_list, count_list = zip(*top_50_words)

# Đọc danh sách stopwords từ tệp
with open('/content/drive/MyDrive/MXH/Stopword.txt', 'r') as stopword_file:
    stopwords1 = stopword_file.read().splitlines()

# Lọc ra các từ stopwords khỏi danh sách word_list
filtered_word_list = [word for word in word_list if word not in stopwords1]

# Tạo biểu đồ cho các từ đã lọc
plt.figure(figsize=(12, 6))
plt.bar(filtered_word_list, [word_counts[word] for word in filtered_word_list])
plt.title('Top 50 Most Common Words (Excluding Stopwords)')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

```

```python
# Đọc tệp res.txt và lấy danh sách bình luận
with open('/content/drive/MyDrive/MXH/res.txt', 'r') as file:
    comments = file.readlines()

# Từ hoặc cụm từ bạn muốn tìm kiếm
search_term = "Love"

# Duyệt qua từng bình luận và kiểm tra xem có chứa từ hoặc cụm từ cần tìm không
found_comments = []
for comment in comments:
    if search_term in comment:
        found_comments.append(comment)

# In ra các bình luận chứa từ hoặc cụm từ cần tìm
for i, comment in enumerate(found_comments, start=1):
    print(f"Bình luận {i}: {comment}")


```

```python
#Cach 1 : PhantichbangTextBlob
from textblob import TextBlob

# Đọc tệp res.txt và lấy danh sách bình luận
with open('/content/drive/MyDrive/MXH/res.txt', 'r') as file:
    comments = file.readlines()

# Đánh giá từng bình luận
for comment in comments:
    # Sử dụng TextBlob để phân tích tình cảm của bình luận
    analysis = TextBlob(comment)

    # Xác định tính cảm của bình luận
    if analysis.sentiment.polarity > 0:
        sentiment = "positive"
    elif analysis.sentiment.polarity < 0:
        sentiment = "negative"
    else:
        sentiment = "neutral"

    print(f"Bình luận: {comment}")
    print(f"Trang thai: {sentiment}\n")
```

```python
#Cach 2 : PhantichbangVader
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Khởi tạo đối tượng SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Đọc dữ liệu từ tệp res.txt
with open('/content/drive/MyDrive/MXH/res.txt', 'r', encoding='utf-8') as file:
    comments = file.readlines()

# Phân tích cảm xúc cho từng comment và in ra kết quả
for comment in comments:
    sentiment = analyzer.polarity_scores(comment)
    print(f"Comment: {comment.strip()}")
    print(f"Sentiment: {sentiment}")
    print("\n")
```
